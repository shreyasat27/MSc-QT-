**Rabi Oscillations**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
delta = np.linspace(-10, 10, 500)  # Detuning range
lambda_vals = np.linspace(-10, 10, 500)  # Coupling range

# Create a meshgrid for delta and lambda
delta_grid, lambda_grid = np.meshgrid(delta, lambda_vals)

# Calculate the Rabi frequency (Omega_R)
Omega_R = 0.5 * np.sqrt(delta_grid**2 + lambda_grid**2)

# Define time (t) and normalize the base frequency for simplicity
omega = 1  # Base frequency for simplicity
t = 1  # Time parameter (adjustable for dynamics)

# Calculate |P1|^2: Probability amplitude squared
P1_squared = (lambda_grid**2 * np.sin(Omega_R * t)**2) / (4 * Omega_R)  # Eq. (137)

# Plotting the result
fig, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(delta_grid, lambda_grid, P1_squared, levels=50, cmap='viridis')
cbar = plt.colorbar(contour)
cbar.set_label('|P1|^2')

# Adding labels and title
ax.set_title('Plot of $|P1|^2$ as a function of $\delta$ and $\lambda$')
ax.set_xlabel('$\delta$ (Detuning)')
ax.set_ylabel('$\lambda$ (Coupling Strength)')
plt.show()

```
