```python
import numpy as np
import matplotlib.pyplot as plt

# Define flux range
phi = np.linspace(-1.2, 1.2, 400)
phi0 = 1  # Flux quantum

# Compute Josephson inductance ratio (green plot)
L_J0 = 1  # Normalized
L_SQ = L_J0 / (2 * np.abs(np.cos(np.pi * phi / phi0)))
L_SQ_norm = 2 * L_SQ / L_J0

# Compute characteristic frequency ratio (red plot)
f0 = np.abs(np.cos(np.pi * phi / phi0))  # Normalized

# Create two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left plot: Inductance
axes[0].plot(phi / phi0, L_SQ_norm, 'g', linewidth=2)
axes[0].set_xlabel(r'$\Phi / \Phi_0$')
axes[0].set_ylabel(r'$2L_{SQ}(\Phi) / L_J$')
axes[0].set_title("Flux Dependence of Josephson Inductance")
axes[0].grid()

# Right plot: Frequency
axes[1].plot(phi / phi0, f0, 'r', linewidth=2)
axes[1].set_xlabel(r'$\Phi / \Phi_0$')
axes[1].set_ylabel(r'$f_0(\Phi) / f_0(0)$')
axes[1].set_title("Flux Dependence of Characteristic Frequency")
axes[1].grid()

plt.tight_layout()
plt.show()
```
