**1. Aharonov-Casher Phase: Path Dependence**

This code simulates a spin-1/2 particle moving along a circular trajectory around a charged line at different tilt angles (Œ∏). It shows how the AC phase changes as ùúÉ varies, confirming the geometric nature of the phase.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
theta_vals = np.linspace(0, np.pi/2, 5)  # Different tilt angles (0 to 90 degrees)
lambda_ratio = np.linspace(0, 3, 100)  # Charge density ratio Œª/Œª‚ÇÄ

# Compute Aharonov-Casher phase for different angles
def ac_phase(lambda_ratio, theta):
    return np.pi * lambda_ratio * np.cos(theta)  # Approximate dependence

plt.figure(figsize=(8,6))
for theta in theta_vals:
    phase = ac_phase(lambda_ratio, theta)
    plt.plot(lambda_ratio, phase, label=f"$\\theta = {np.degrees(theta):.1f}^\circ$")

plt.xlabel(r"Charge Density Ratio $\lambda / \lambda_0$")
plt.ylabel(r"Aharonov-Casher Phase $\phi_{AC}$ (radians)")
plt.title("Path-Dependent Aharonov-Casher Phase")
plt.legend()
plt.grid()
plt.show()
```

**Expected Plot:**

A family of curves showing how the AC phase changes with charge density.
When
Œ∏=0, the phase depends only on charge density (suggesting topological behavior).

For ùúÉ>0, the phase changes, proving path dependence.

**2. Scalar Aharonov-Bohm Effect: Path Dependence**

This code simulates a charged particle moving through regions with different electric potentials and calculates the phase accumulation along different paths.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define spatial coordinate (path variable)
x = np.linspace(-2, 2, 200)
V1 = 1 / (1 + np.exp(-5*x))  # Electric potential along path 1
V2 = np.exp(-x**2)  # Potential along path 2

# Compute the phase accumulated along each path
phi_path1 = np.cumsum(V1) * (x[1] - x[0])  # Integral approximation
phi_path2 = np.cumsum(V2) * (x[1] - x[0])

plt.figure(figsize=(8,6))
plt.plot(x, phi_path1, label="Path 1 (Steep potential change)", lw=2)
plt.plot(x, phi_path2, label="Path 2 (Smooth potential change)", lw=2, linestyle="dashed")

plt.xlabel("Position (x)")
plt.ylabel("Accumulated Phase $\phi_{AB}^{scalar}$")
plt.title("Path-Dependent Scalar Aharonov-Bohm Phase")
plt.legend()
plt.grid()
plt.show()
```

**Expected Plot:**

Two different phase accumulation curves for different potential profiles.
The phase is not just dependent on the total potential difference but also on how the potential varies along the path.

**3. Pancharatnam-Berry Phase in Optics**

This code visualizes the Pancharatnam phase by simulating the evolution of polarization states on the Poincar√© sphere.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate Poincar√© sphere
theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2*np.pi, 100)
Theta, Phi = np.meshgrid(theta, phi)
X = np.sin(Theta) * np.cos(Phi)
Y = np.sin(Theta) * np.sin(Phi)
Z = np.cos(Theta)

# Define a polarization path (cyclic evolution)
phi_path = np.linspace(0, 2*np.pi, 50)
theta_path = np.pi/4 * np.ones_like(phi_path)
X_path = np.sin(theta_path) * np.cos(phi_path)
Y_path = np.sin(theta_path) * np.sin(phi_path)
Z_path = np.cos(theta_path)

# Plot the Poincar√© sphere
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='cyan', alpha=0.3)
ax.plot(X_path, Y_path, Z_path, 'r', lw=2, label="Polarization Evolution")

ax.set_xlabel("S1 (Horizontal)")
ax.set_ylabel("S2 (Diagonal)")
ax.set_zlabel("S3 (Circular)")
ax.set_title("Pancharatnam Phase Evolution on the Poincar√© Sphere")
plt.legend()
plt.show()
```

**Expected Plot:**

A 3D sphere (Poincar√© sphere) representing polarization states.
A red cyclic path showing the geometric evolution of polarization.
The enclosed solid angle represents the Pancharatnam phase.
