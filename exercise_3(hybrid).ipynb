```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Create figure and axes
fig, ax = plt.subplots(figsize=(10, 3))
ax.set_xlim(0, 10)
ax.set_ylim(0, 3)
ax.axis('off')

# Draw Source
source = patches.Rectangle((0.5, 1.2), 0.6, 0.6, facecolor='gray')
ax.add_patch(source)
ax.text(0.8, 1.85, 'Source\n(FM)', ha='center')

# Draw Drain
drain = patches.Rectangle((8.9, 1.2), 0.6, 0.6, facecolor='gray')
ax.add_patch(drain)
ax.text(9.2, 1.85, 'Drain\n(FM)', ha='center')

# Draw 2DEG channel
channel = patches.Rectangle((1.1, 1.4), 7.8, 0.2, facecolor='lightblue')
ax.add_patch(channel)
ax.text(5, 1.7, '2DEG channel', ha='center', fontsize=10)

# Draw Gate
gate = patches.Rectangle((3.2, 2.1), 3.5, 0.3, facecolor='lightcoral')
ax.add_patch(gate)
ax.text(5, 2.5, 'Gate (controls SOC)', ha='center', fontsize=10)

# Draw spin precession arrows
x_vals = np.linspace(1.3, 8.3, 7)
for i, x in enumerate(x_vals):
    angle = i * np.pi / 6
    dx = 0.2 * np.cos(angle)
    dy = 0.2 * np.sin(angle)
    ax.arrow(x, 1.5, dx, dy, head_width=0.05, head_length=0.05, color='purple')

# Add spin labels
ax.text(1.2, 1.1, '↑', fontsize=14)
ax.text(8.4, 1.1, '↓', fontsize=14)

plt.title('Schematic of a Datta-Das Spin FET', fontsize=14)
plt.tight_layout()
plt.show()
```
